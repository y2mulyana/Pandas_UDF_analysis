## UDF section

1. UDF for continouos data

```python
def continuous_summary(x):
  return pd.Series([x.count(), x.sum(), x.min(), x.median(), x.std(), x.var(), x.quantile(0.01), x.quantile(0.25), x.quantile(0.5), x.quantile(0.75), x.quantile(0.99), x.max()],
                   index = ['N', 'SUM', 'MIN', 'MEDIAN', 'STDEV', 'VAR', 'P1', 'P25', 'P50', 'P75', 'P99', 'MAX'])
```

```python
df_cont_vars.apply(continuous_summary).T.round(2)
```

2. UDF for categorycal data

```python
def categorycal_summary(y):
  Mode = y.value_counts().reset_index()
  return pd.Series([y.count(), y.isna().sum(), Mode.iloc[0,0], Mode.iloc[0, 1], (Mode.iloc[0,1]*100 / y.count())],
                   index= ['N', 'N-MISS', 'MODE', 'FREQ', 'PERC'])
```

```python
df_cat_vars.apply(categorycal_summary).T.round(2)
```
